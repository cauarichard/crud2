```python
# COMPLETE - Extração + t-SNE num código só
import os, numpy as np, shutil
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing import image
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')
base_dir = '/content/drive/MyDrive/dataset/real_fake/faces_geral_real_fake'
NUM_GRUPOS = 5

# EXTRACAO
model = VGG16(weights='imagenet', include_top=False)
file_list, feature_list = [], []
for file_name in os.listdir(base_dir):
    if file_name.lower().endswith(('.png', '.jpg', '.jpeg')):
        img_path = os.path.join(base_dir, file_name)
        try:
            img = image.load_img(img_path, target_size=(224, 224))
            img_data = image.img_to_array(img); img_data = np.expand_dims(img_data, 0)
            img_data = preprocess_input(img_data)
            features = model.predict(img_data, verbose=0).flatten()
            file_list.append(file_name); feature_list.append(features)
        except: pass

# K-MEANS
kmeans = KMeans(n_clusters=NUM_GRUPOS, random_state=42)
grupos = kmeans.fit_predict(np.array(feature_list))

# T-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
features_2d = tsne.fit_transform(np.array(feature_list))

# Labels real/fake
labels = ['Real' if 'real' in f.lower() else 'Fake' if 'fake' in f.lower() else 'Outros'
          for f in file_list]

# PLOT
df = pd.DataFrame({'x': features_2d[:,0], 'y': features_2d[:,1], 'grupo': grupos, 'classe': labels})
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.scatterplot(data=df, x='x', y='y', hue='classe', s=50); plt.title('t-SNE Real vs Fake')
plt.subplot(1,2,2)
sns.scatterplot(data=df, x='x', y='y', hue='grupo', palette='husl', s=50); plt.title('t-SNE K-Means')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/tsne_vgg16.png', dpi=300, bbox_inches='tight')
plt.show()

print(df.groupby(['grupo','classe']).size())
```
